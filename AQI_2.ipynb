#### Here we have taken the data from another source. We wanted values of 'PM2.5'(It is a parameter for measurement of AQI) which was not available in the previous data(it is all average values per day).

#### Here the data which we are taking from other source is  PER HOUR data. We will convert it to PER DAY because all the other data which we have is PER DAY

```python
import pandas as pd
import matplotlib.pyplot as plt
```

#### Computing average for PM2.5 for each day in a particular year (2013)

```python
def avg_data_2013():
    temp_i=0 #counter
    average=[]
    for rows in pd.read_csv('Data/AQI/aqi2013.csv',chunksize=24):#to read csv in a batch of 24(24 hours we have) rows at a time
        add_var = 0 #to add every hour data
        avg = 0.0
        data = []
        df = pd.DataFrame(data = rows)#converting the 24 rows into dataframe
        for index,row in df.iterrows():#iterating over the 24 rows
            data.append(row['PM2.5'])#fetching the value of the column:'PM2.5' from each rowin df and appending to an empty list
        for i in data:#iterating over the list
            if type(i) is float or type(i) is int:
                add_var = add_var + i
            elif type(i) is str:#changing the data_type of data from str to float for calculation purpose
                if i!='NoData' and i!='PwrFail' and i!='---' and i!='InVld':#handling the 'INVALID' data.We are just ignoring it
                    temp = float(i)
                    add_var = add_var + temp
        avg = add_var/24
        temp_i = temp_i + 1
        
        average.append(avg)
    return average    
```

```python
#%%time
if __name__ == '__main__':
    lst2013 = avg_data_2013()
    print(lst2013)
```

#### Computing average for PM2.5 for each day for ALL YEARS from 2013 to 2018

```python
def avg_data():
    filenames = glob('Data\AQI\*.csv')
    dataframes = [f for f in filenames]
    avg_list_for_all_years = []
    for f in filenames:
        temp_i=0
        average=[]
        for rows in pd.read_csv(f,chunksize = 24):
            add_var = 0
            avg = 0.0
            data = []
            df = pd.DataFrame(data = rows)
            for index,row in df.iterrows():
                data.append(row['PM2.5'])
            for i in data:
                if type(i) is float or type(i) is int:
                    add_var = add_var + i
                elif type(i) is str:
                    if i!='NoData' and i!='PwrFail' and i!='---' and i!='InVld':
                        temp = float(i)
                        add_var = add_var + temp
                avg = add_var/24
                temp_i = temp_i + 1
        
            average.append(avg)
        avg_list_for_all_years.append(average)
    return avg_list_for_all_years
```

```python
#%%time
if __name__ == '__main__':
    lst_all_years = avg_data()
    print(lst_all_years)
```
